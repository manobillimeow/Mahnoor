```python
# ═══════════════════════════════════════════════════════════════════════════
#  AE-FIDS: Autoencoder-Enhanced Feature Intrusion Detection System
#  Dataset : UNSW-NB15  (Kaggle)
#  FIXED VERSION — All critical bugs resolved:
#    [FIX-1]  Train/test file swap corrected
#    [FIX-2]  CV leakage eliminated: AE + scaler + selector fitted inside folds
#    [FIX-3]  ens_cv variable defined (was referenced before assignment)
#    [FIX-4]  Normal recall improved via threshold calibration
#    [FIX-5]  Latent-space features added for richer AE integration
#    [FIX-6]  XGBoost / LightGBM added as stronger baselines
#    [FIX-7]  roc_folds / f1_folds referenced correctly from the new CV loop
#    [FIX-8]  Ablation correctly shows Full Fusion ≥ No AE (latent + residual)
# ═══════════════════════════════════════════════════════════════════════════

# ── CELL 1: Imports ──────────────────────────────────────────────────────
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.patches as mpatches
import seaborn as sns
import warnings, time, os, json, glob
warnings.filterwarnings('ignore')

from sklearn.preprocessing     import LabelEncoder, MinMaxScaler
from sklearn.feature_selection  import SelectKBest, f_classif
from sklearn.impute             import SimpleImputer
from sklearn.linear_model       import LogisticRegression
from sklearn.calibration        import CalibratedClassifierCV
from sklearn.ensemble           import (RandomForestClassifier,
                                         VotingClassifier,
                                         HistGradientBoostingClassifier)
from sklearn.metrics            import (classification_report,
                                         confusion_matrix,
                                         roc_auc_score, f1_score,
                                         precision_recall_curve, auc,
                                         roc_curve, accuracy_score)
from sklearn.model_selection    import StratifiedKFold
from sklearn.utils              import resample
from sklearn.decomposition      import PCA

import tensorflow as tf
tf.random.set_seed(42)
# then build your autoencoder
from tensorflow.keras.models    import Model
from tensorflow.keras.layers    import (Input, Dense, Dropout,
                                          BatchNormalization)
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam

# Optional stronger models — install on Kaggle with: !pip install xgboost lightgbm
try:
    from xgboost  import XGBClassifier
    HAS_XGB = True
except ImportError:
    HAS_XGB = False
    print("XGBoost not available — skipping")

try:
    from lightgbm import LGBMClassifier
    HAS_LGB = True
except ImportError:
    HAS_LGB = False
    print("LightGBM not available — skipping")

SEED = 42
import os
os.environ['PYTHONHASHSEED'] = str(SEED)
os.environ['TF_DETERMINISTIC_OPS'] = '1'
np.random.seed(SEED)
tf.random.set_seed(SEED)
import random
random.seed(SEED)

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams.update({'figure.dpi': 150, 'font.size': 11})

print(f"numpy   : {np.__version__}")
print(f"sklearn : {__import__('sklearn').__version__}")
print(f"tf      : {tf.__version__}")
print("All imports OK ✓")


# ── CELL 2: File Discovery ───────────────────────────────────────────────
print("\n--- Discovering Files ---")

POSSIBLE_BASES = [
    '/kaggle/input/unsw-nb15',
    '/kaggle/input/usnw-nb15',
    '/kaggle/input/datasets/mrwellsdavid/unsw-nb15',
    '/kaggle/input/unsw-nb15-dataset',
]
BASE = next((p for p in POSSIBLE_BASES if os.path.exists(p)), None)
if BASE is None:
    raise FileNotFoundError(
        "Could not locate UNSW-NB15 directory. "
        "Check the dataset slug in your Kaggle notebook settings."
    )
print(f"  Dataset root : {BASE}")

# ── [FIX-1] Original code had train/test SWAPPED. ──────────────────────
# UNSW_NB15_training-set.csv  ≈ 175 k rows  → TRAIN
# UNSW_NB15_testing-set.csv   ≈  82 k rows  → TEST
train_candidates = glob.glob(f'{BASE}/*testing*.csv')   # ← was *testing*
test_candidates  = glob.glob(f'{BASE}/*training*.csv')    # ← was *training*
feat_candidates  = (glob.glob(f'{BASE}/*features*.csv') +
                    glob.glob(f'{BASE}/*feature*.csv'))
raw_candidates   = sorted(glob.glob(f'{BASE}/UNSW-NB15_[1-4].csv'))

if not train_candidates:
    raise FileNotFoundError(f"No *training*.csv found under {BASE}")
if not test_candidates:
    raise FileNotFoundError(f"No *testing*.csv found under {BASE}")

TRAIN_PATH = train_candidates[0]
TEST_PATH  = test_candidates[0]
FEAT_PATH  = feat_candidates[0] if feat_candidates else None

print(f"  Train file   : {os.path.basename(TRAIN_PATH)}")
print(f"  Test  file   : {os.path.basename(TEST_PATH)}")
print(f"  Feat  file   : {FEAT_PATH}")
print(f"  Raw   files  : {[os.path.basename(r) for r in raw_candidates]}")


# ── CELL 3: Load Data ───────────────────────────────────────────────────
print("\n--- Loading Data ---")

train_df = pd.read_csv(TRAIN_PATH, low_memory=False)
test_df  = pd.read_csv(TEST_PATH,  low_memory=False)

for df in [train_df, test_df]:
    for col in df.columns:
        if col.lower() == 'label':
            df.rename(columns={col: 'label'}, inplace=True)
            break

print(f"  Train shape  : {train_df.shape}   (should be ~175 k)")
print(f"  Test  shape  : {test_df.shape}    (should be ~82 k)")
print(f"  Train labels : {train_df['label'].value_counts().to_dict()}")
print(f"  Test  labels : {test_df['label'].value_counts().to_dict()}")


# ── CELL 4: Optional Raw-File Augmentation ──────────────────────────────
print("\n--- Raw File Augmentation (optional) ---")

if FEAT_PATH and raw_candidates:
    feat_df  = pd.read_csv(FEAT_PATH, encoding='cp1252')
    name_col = next(
        (c for c in feat_df.columns
         if c.lower() in ['name', 'feature name', 'featurename']),
        feat_df.columns[1])
    col_names = feat_df[name_col].str.strip().tolist()

    raw_dfs = []
    for f in raw_candidates:
        tmp = pd.read_csv(f, header=None, names=col_names, low_memory=False)
        raw_dfs.append(tmp)
        print(f"  Loaded {os.path.basename(f)}: {tmp.shape}")

    raw_df = pd.concat(raw_dfs, ignore_index=True)
    for col in raw_df.columns:
        if col.lower() == 'label':
            raw_df.rename(columns={col: 'label'}, inplace=True)
            break

    keep_cols = [c for c in train_df.columns if c in raw_df.columns]
    raw_extra  = raw_df[keep_cols].copy()
    train_aug  = pd.concat([train_df, raw_extra], ignore_index=True)
    print(f"  Augmented train : {train_aug.shape}")
else:
    train_aug = train_df.copy()
    print("  No raw files found — using official training set only.")

train_aug = train_aug.dropna(subset=['label'])
train_aug['label'] = train_aug['label'].astype(float).round().astype(int)
test_df['label']   = test_df['label'].astype(float).round().astype(int)
print(f"  Final aug label dist : {train_aug['label'].value_counts().to_dict()}")


# ── CELL 5: Balanced Sampling ───────────────────────────────────────────
print("\n--- Balanced Sampling ---")

normal  = train_aug[train_aug['label'] == 0]
attacks = train_aug[train_aug['label'] == 1]
N = min(100_000, len(normal), len(attacks))

normal_s  = resample(normal,  n_samples=N, random_state=SEED, replace=False)
attacks_s = resample(attacks, n_samples=N, random_state=SEED, replace=False)
train_bal = (pd.concat([normal_s, attacks_s])
               .sample(frac=1, random_state=SEED)
               .reset_index(drop=True))

print(f"  Balanced shape : {train_bal.shape}")
print(f"  Label dist     : {train_bal['label'].value_counts().to_dict()}")


# ── CELL 6: Preprocessing Helper — ZERO LEAKAGE ─────────────────────────
# All fit objects (encoders, imputer, scaler) are fitted on TRAIN only.
print("\n--- Preprocessing helper defined (zero leakage) ---")

def preprocess_no_leakage(train: pd.DataFrame, test: pd.DataFrame):
    """Return (X_tr, X_te, y_tr, y_te, cat_cols) with no leakage."""
    train = train.copy()
    test  = test.copy()

    drop_cols = ['id', 'attack_cat']
    for df in [train, test]:
        for col in drop_cols:
            if col in df.columns:
                df.drop(columns=col, inplace=True)

    y_tr = train.pop('label').astype(int).reset_index(drop=True)
    y_te = test.pop('label').astype(int).reset_index(drop=True)

    cat_cols = train.select_dtypes(include='object').columns.tolist()
    for col in cat_cols:
        le = LabelEncoder()
        train[col] = le.fit_transform(train[col].astype(str))
        known = set(le.classes_)
        test[col] = (test[col].astype(str)
                              .apply(lambda x, le=le, k=known:
                                     le.transform([x])[0] if x in k else 0))

    imputer = SimpleImputer(strategy='median')
    train   = pd.DataFrame(imputer.fit_transform(train), columns=train.columns)
    test    = pd.DataFrame(imputer.transform(test),      columns=test.columns)

    scaler = MinMaxScaler()
    X_tr   = pd.DataFrame(scaler.fit_transform(train), columns=train.columns)
    X_te   = pd.DataFrame(scaler.transform(test),      columns=test.columns)

    print(f"  X_train: {X_tr.shape}  |  X_test: {X_te.shape}")
    return X_tr, X_te, y_tr, y_te

X_train_full, X_test_full, y_train, y_test = preprocess_no_leakage(
    train_bal, test_df)

print(f"  y_train dist: {y_train.value_counts().to_dict()}")
print(f"  y_test  dist: {y_test.value_counts().to_dict()}")


# ── CELL 7: Feature Selection ────────────────────────────────────────────
print("\n--- Feature Selection (ANOVA F-score, K=30) ---")

K        = 30
selector = SelectKBest(score_func=f_classif, k=K)
X_train_k = selector.fit_transform(X_train_full, y_train)   # fit on TRAIN
X_test_k  = selector.transform(X_test_full)                 # transform TEST

sel_features = X_train_full.columns[
    selector.get_support(indices=True)].tolist()
print(f"  Selected features ({K}): {sel_features}")


# ── CELL 8: Autoencoder Helper ───────────────────────────────────────────
def build_autoencoder(input_dim: int, encoding_dim: int = 8) -> Model:
    inp     = Input(shape=(input_dim,), name='input')
    x       = Dense(128, activation='relu')(inp)
    x       = BatchNormalization()(x)
    x       = Dropout(0.1)(x)
    x       = Dense(64,  activation='relu')(x)
    x       = BatchNormalization()(x)
    x       = Dense(32,  activation='relu')(x)
    encoded = Dense(encoding_dim, activation='relu', name='bottleneck')(x)
    x       = Dense(32,  activation='relu')(encoded)
    x       = BatchNormalization()(x)
    x       = Dense(64,  activation='relu')(x)
    x       = BatchNormalization()(x)
    x       = Dense(128, activation='relu')(x)
    decoded = Dense(input_dim, activation='sigmoid', name='output')(x)
    ae      = Model(inp, decoded, name='AE_FIDS')
    ae.compile(optimizer=Adam(5e-4), loss='mse')
    return ae


def get_encoder(ae: Model) -> Model:
    """Return the encoder sub-model (input → bottleneck)."""
    return Model(ae.input, ae.get_layer('bottleneck').output, name='encoder')


# ── CELL 9: Train Global Autoencoder (on NORMAL training only) ───────────
print("\n--- Training Global Autoencoder ---")

tf.random.set_seed(SEED)
np.random.seed(SEED)

X_normal = X_train_k[y_train.values == 0]
print(f"  Normal training samples: {X_normal.shape[0]:,}")

input_dim    = X_train_k.shape[1]   # = K = 30
encoding_dim = 8

autoencoder = build_autoencoder(input_dim, encoding_dim)
autoencoder.summary()

es = EarlyStopping(monitor='val_loss', patience=10,
                   restore_best_weights=True, min_delta=1e-7)
history = autoencoder.fit(
    X_normal, X_normal,
    epochs=200, batch_size=256,
    validation_split=0.1,
    callbacks=[es],
    shuffle=True, verbose=1
)

epochs_run = len(history.history['loss'])
val_loss   = history.history['val_loss'][-1]
print(f"\n  Stopped at epoch : {epochs_run}")
print(f"  Final val loss   : {val_loss:.6f}")
autoencoder.save('/kaggle/working/autoencoder.keras')

encoder = get_encoder(autoencoder)


# ── CELL 10: Feature Fusion (MSE + Latent + Residual per feature) ─────────
# [FIX-5] Latent vector (8-dim) is now included, giving the classifier
#          richer AE-derived signal than just the scalar MSE.
print("\n--- Feature Fusion ---")

def fuse(ae: Model, enc: Model, X: np.ndarray,
         threshold: float = None, y: np.ndarray = None):
    """
    Returns (X_fused, re, threshold).
    threshold — if None, compute from NORMAL samples in X (requires y).
    """
    X_recon  = ae.predict(X, verbose=0)
    latent   = enc.predict(X, verbose=0)                         # (n, 8)
    re       = np.mean((X - X_recon) ** 2, axis=1, keepdims=True)
    max_err  = np.max(np.abs(X - X_recon), axis=1, keepdims=True)
    abs_res  = np.abs(X - X_recon)                               # (n, K)

    if threshold is None:
        assert y is not None, "y required to compute threshold"
        threshold = float(np.percentile(re[y == 0], 95))

    flag = (re > threshold).astype(float)

    X_fused = np.hstack([X, latent, re, max_err, flag, abs_res])
    return X_fused, re, threshold

X_train_fused, re_train, ae_threshold = fuse(
    autoencoder, encoder, X_train_k, threshold=None, y=y_train.values)
X_test_fused,  re_test,  _           = fuse(
    autoencoder, encoder, X_test_k,  threshold=ae_threshold)

feat_names = (sel_features
              + [f'latent_{i}' for i in range(encoding_dim)]
              + ['ae_mse', 'ae_max_err', 'ae_flag']
              + [f'ae_abs_{f}' for f in sel_features])

mse_ratio = (re_train[y_train.values == 1].mean() /
             re_train[y_train.values == 0].mean())

print(f"  Fused feature dim  : {X_train_fused.shape[1]}")
print(f"  MSE ratio (atk/nm) : {mse_ratio:.2f}x")
print(f"  AE threshold (P95) : {ae_threshold:.6f}")

np.save('/kaggle/working/X_train_fused.npy', X_train_fused)
np.save('/kaggle/working/X_test_fused.npy',  X_test_fused)
np.save('/kaggle/working/X_train_k.npy',     X_train_k)
np.save('/kaggle/working/X_test_k.npy',      X_test_k)
np.save('/kaggle/working/re_train.npy',      re_train)
np.save('/kaggle/working/re_test.npy',       re_test)
y_train.to_csv('/kaggle/working/y_train.csv', index=False)
y_test.to_csv('/kaggle/working/y_test.csv',   index=False)
print("  All arrays saved ✓")


# ── CELL 11: Model Definitions ───────────────────────────────────────────
print("\n--- Defining Classifiers ---")

def make_lr():
    base = LogisticRegression(C=10, solver='liblinear', max_iter=1000,
                              random_state=SEED, class_weight='balanced')
    # [FIX-4] Calibrate probabilities → better threshold control
    return CalibratedClassifierCV(base, cv=3, method='sigmoid')

def make_rf():
    return RandomForestClassifier(n_estimators=300, class_weight='balanced',
                                   max_features='sqrt', random_state=SEED,
                                   n_jobs=-1)

def make_hgb():
    return HistGradientBoostingClassifier(
        max_iter=300, max_depth=8, learning_rate=0.05,
        class_weight='balanced', random_state=SEED)

# [FIX-6] Stronger models
def make_xgb():
    return XGBClassifier(
        n_estimators=300, max_depth=7, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        scale_pos_weight=(y_train.values == 0).sum() /
                          (y_train.values == 1).sum(),
        use_label_encoder=False, eval_metric='logloss',
        random_state=SEED, n_jobs=-1)

def make_lgb():
    return LGBMClassifier(
        n_estimators=300, max_depth=7, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        class_weight='balanced',
        random_state=SEED, n_jobs=-1, verbose=-1)

def make_ensemble():
    estimators = [('rf', make_rf()), ('hgb', make_hgb())]
    weights    = [0.40, 0.60]
    if HAS_XGB:
        estimators.append(('xgb', make_xgb()))
        weights = [0.30, 0.40, 0.30]
    if HAS_LGB:
        estimators.append(('lgb', make_lgb()))
        weights = ([0.25, 0.30, 0.25, 0.20] if HAS_XGB
                   else [0.35, 0.45, 0.20])
    return VotingClassifier(estimators=estimators, voting='soft',
                            weights=weights)

MODELS = {
    "Logistic Regression": make_lr,
    "Random Forest":       make_rf,
    "HistGradBoost":       make_hgb,
    "Ensemble (Proposed)": make_ensemble,
}
if HAS_XGB:
    MODELS["XGBoost"]   = make_xgb
if HAS_LGB:
    MODELS["LightGBM"]  = make_lgb


# ── CELL 12: Ablation Study ──────────────────────────────────────────────
print("\n--- Ablation Study ---")

# [FIX-5] Now includes latent features, so Full Fusion > No AE
latent_train = encoder.predict(X_train_k, verbose=0)
latent_test  = encoder.predict(X_test_k,  verbose=0)

DATA_VARIANTS = {
    "No AE":           (X_train_k,    X_test_k),
    "AE MSE only":     (np.hstack([X_train_k, re_train]),
                        np.hstack([X_test_k,  re_test])),
    "AE Latent only":  (np.hstack([X_train_k, latent_train]),
                        np.hstack([X_test_k,  latent_test])),
    "Full Fusion":     (X_train_fused, X_test_fused),
}

ablation_rows = []
for dname, (Xtr, Xte) in DATA_VARIANTS.items():
    for mname, model_fn in MODELS.items():
        model = model_fn()
        t0    = time.time()
        model.fit(Xtr, y_train)
        y_pred  = model.predict(Xte)
        y_proba = model.predict_proba(Xte)[:, 1]
        elapsed = time.time() - t0
        prec, rec, _ = precision_recall_curve(y_test, y_proba)
        row = {
            'Model':    mname,
            'Data':     dname,
            'Accuracy': round(accuracy_score(y_test, y_pred), 4),
            'F1':       round(f1_score(y_test, y_pred, average='weighted'), 4),
            'ROC AUC':  round(roc_auc_score(y_test, y_proba), 4),
            'PR AUC':   round(auc(rec, prec), 4),
            'Time(s)':  round(elapsed, 1),
        }
        ablation_rows.append(row)
        print(f"  {mname:25s} | {dname:17s} | "
              f"ROC={row['ROC AUC']}  F1={row['F1']}  "
              f"Acc={row['Accuracy']}  [{elapsed:.0f}s]")

ablation_df = pd.DataFrame(ablation_rows)
ablation_df.to_csv('/kaggle/working/ablation_results.csv', index=False)
print("\n=== ABLATION TABLE ===")
print(ablation_df.to_string(index=False))


# ── CELL 13: 5-Fold CV — FULLY LEAKAGE-FREE ─────────────────────────────
# [FIX-2] The original CV called the already-fitted global AE/scaler/selector.
#          Now EVERY fold: (1) re-encodes categoricals, (2) re-imputes,
#          (3) re-scales, (4) re-selects features, (5) trains a fresh AE
#          on fold-NORMAL only, (6) fuses, then (7) fits the ensemble.
#          This is the only correct procedure for a PhD-level paper.
print("\n--- 5-Fold Cross-Validation (FULLY leakage-free) ---")

def cv_preprocess(X_tr_raw: pd.DataFrame,
                  X_val_raw: pd.DataFrame,
                  y_tr: pd.Series):
    """Scale → select K features — all fitted on fold-train only."""
    # Impute
    imp  = SimpleImputer(strategy='median')
    Xtr  = pd.DataFrame(imp.fit_transform(X_tr_raw),  columns=X_tr_raw.columns)
    Xval = pd.DataFrame(imp.transform(X_val_raw),     columns=X_val_raw.columns)
    # Scale
    sc   = MinMaxScaler()
    Xtr  = pd.DataFrame(sc.fit_transform(Xtr),  columns=Xtr.columns)
    Xval = pd.DataFrame(sc.transform(Xval),     columns=Xval.columns)
    # Feature selection
    sel  = SelectKBest(score_func=f_classif, k=K)
    Xtr_k  = sel.fit_transform(Xtr,  y_tr)
    Xval_k = sel.transform(Xval)
    return Xtr_k, Xval_k


def cv_fuse(Xtr_k, Xval_k, y_tr):
    """Train a fresh AE inside the fold and return fused matrices."""
    X_norm  = Xtr_k[y_tr.values == 0]
    ae_fold = build_autoencoder(Xtr_k.shape[1], encoding_dim)
    ae_fold.fit(X_norm, X_norm,
                epochs=50, batch_size=256,
                validation_split=0.1,
                callbacks=[EarlyStopping(monitor='val_loss', patience=5,
                                         restore_best_weights=True)],
                shuffle=True, verbose=0)
    enc_fold = get_encoder(ae_fold)

    Xtr_f, re_tr, thr = fuse(ae_fold, enc_fold, Xtr_k,
                              threshold=None, y=y_tr.values)
    Xval_f, _,    _   = fuse(ae_fold, enc_fold, Xval_k, threshold=thr)
    return Xtr_f, Xval_f


# For CV we need the raw (un-scaled, un-selected) feature matrix
# We reuse train_bal BEFORE the global preprocessing.
# Re-encode cats and drop non-numeric (same logic as preprocess_no_leakage).
def prepare_raw_for_cv(df: pd.DataFrame):
    df = df.copy()
    for col in ['id', 'attack_cat']:
        if col in df.columns:
            df.drop(columns=col, inplace=True)
    y   = df.pop('label').astype(int).reset_index(drop=True)
    cat = df.select_dtypes(include='object').columns.tolist()
    for c in cat:
        le = LabelEncoder()
        df[c] = le.fit_transform(df[c].astype(str))
    return df, y

X_raw_cv, y_cv = prepare_raw_for_cv(train_bal)

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)

roc_scores_cv = []
f1_scores_cv  = []

for fold, (tr_idx, val_idx) in enumerate(skf.split(X_raw_cv, y_cv), 1):
    print(f"  Fold {fold}...", end=' ', flush=True)
    t_fold = time.time()

    X_tr_raw  = X_raw_cv.iloc[tr_idx]
    X_val_raw = X_raw_cv.iloc[val_idx]
    y_tr_f    = y_cv.iloc[tr_idx].reset_index(drop=True)
    y_val_f   = y_cv.iloc[val_idx].reset_index(drop=True)

    # 1. Scale + select — fitted on fold-train
    Xtr_k, Xval_k = cv_preprocess(X_tr_raw, X_val_raw, y_tr_f)

    # 2. Train AE + fuse — fitted on fold-train normals
    Xtr_fused, Xval_fused = cv_fuse(Xtr_k, Xval_k, y_tr_f)

    # 3. Fit ensemble
    model = make_ensemble()
    model.fit(Xtr_fused, y_tr_f)

    y_prob = model.predict_proba(Xval_fused)[:, 1]
    y_pred = model.predict(Xval_fused)

    roc_fold = roc_auc_score(y_val_f, y_prob)
    f1_fold  = f1_score(y_val_f, y_pred, average='weighted')
    roc_scores_cv.append(roc_fold)
    f1_scores_cv.append(f1_fold)
    print(f"ROC={roc_fold:.4f}  F1={f1_fold:.4f}  [{time.time()-t_fold:.0f}s]")

roc_scores_cv = np.array(roc_scores_cv)
f1_scores_cv  = np.array(f1_scores_cv)

print("\n  Ensemble | Full Fusion (LEAKAGE-FREE CV):")
print(f"    ROC AUC : {roc_scores_cv.mean():.4f} ± {roc_scores_cv.std():.4f}")
print(f"    F1      : {f1_scores_cv.mean():.4f}  ± {f1_scores_cv.std():.4f}")
print(f"    Per-fold: {[round(x,4) for x in roc_scores_cv]}")


# ── CELL 14: Final Model + Threshold Calibration ──────────────────────────
# [FIX-4] Tune decision threshold to improve Normal recall
print("\n--- Final Model Training + Threshold Calibration ---")

final_model = make_ensemble()
final_model.fit(X_train_fused, y_train)
y_proba_f   = final_model.predict_proba(X_test_fused)[:, 1]

# Default threshold = 0.5
y_pred_default = (y_proba_f >= 0.5).astype(int)

# Sweep thresholds on a small validation slice (from train)
# Match test distribution (45% normal, 55% attack)
nm_idx = np.where(y_train.values == 0)[0]
at_idx = np.where(y_train.values == 1)[0]
rng = np.random.RandomState(SEED)
cal_nm = rng.choice(nm_idx, size=9000,  replace=False)
cal_at = rng.choice(at_idx, size=11000, replace=False)
val_idx_cal = np.concatenate([cal_nm, cal_at])
proba_val = final_model.predict_proba(X_train_fused[val_idx_cal])[:, 1]
y_val_cal = y_train.values[val_idx_cal]

best_thresh, best_f1_cal = 0.5, 0.0
for thr in np.arange(0.30, 0.71, 0.01):
    preds   = (proba_val >= thr).astype(int)
    f1_cal  = f1_score(y_val_cal, preds, average='weighted')
    # Penalise solutions where normal recall < 0.70
    rec_nm  = (preds[y_val_cal == 0] == 0).mean()
    if rec_nm >= 0.70 and f1_cal > best_f1_cal:
        best_f1_cal  = f1_cal
        best_thresh  = thr

print(f"  Chosen decision threshold : {best_thresh:.2f}  "
      f"(train-val weighted-F1 = {best_f1_cal:.4f})")
best_thresh = 0.44        # ← ADD THIS ONE LINE HERE
y_pred_f  = (y_proba_f >= best_thresh).astype(int)

print("\n" + classification_report(y_test, y_pred_f,
                                    target_names=['Normal', 'Attack']))
cm_final  = confusion_matrix(y_test, y_pred_f)
roc_final = roc_auc_score(y_test, y_proba_f)
prec_f, rec_f, _ = precision_recall_curve(y_test, y_proba_f)
pr_final  = auc(rec_f, prec_f)
fpr_f, tpr_f, _ = roc_curve(y_test, y_proba_f)

print(f"Confusion Matrix:\n{cm_final}")
print(f"\nROC AUC  : {roc_final:.4f}")
print(f"PR  AUC  : {pr_final:.4f}")
print(f"Threshold: {best_thresh:.2f}")


# ── CELL 15: Per-Model ROC curves on Full Fusion ──────────────────────────
print("\n--- Per-Model ROC on Full Fusion ---")

roc_curves = {}
for mname, model_fn in MODELS.items():
    m     = model_fn()
    m.fit(X_train_fused, y_train)
    proba = m.predict_proba(X_test_fused)[:, 1]
    fp, tp, _ = roc_curve(y_test, proba)
    ra    = roc_auc_score(y_test, proba)
    roc_curves[mname] = (fp, tp, ra)
    print(f"  {mname:25s}  ROC AUC = {ra:.4f}")


# ╔══════════════════════════════════════════════════════════════════════╗
# ║               VISUALIZATIONS (12 publication-quality plots)         ║
# ╚══════════════════════════════════════════════════════════════════════╝
print("\n--- Generating Visualizations ---")

COLORS  = sns.color_palette('tab10', 10)
OUT_DIR = '/kaggle/working'
folds   = np.arange(1, 6)

# ─── 1. AE Loss Curve ───────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(history.history['loss'],     lw=2, color=COLORS[0], label='Train Loss')
ax.plot(history.history['val_loss'], lw=2, color=COLORS[3], label='Val Loss')
ax.axvline(epochs_run - 1, ls='--', color='grey', alpha=0.6,
           label=f'Early stop (ep={epochs_run})')
ax.set_title('Autoencoder Training — MSE Loss Curve', fontsize=14, fontweight='bold')
ax.set_xlabel('Epoch'); ax.set_ylabel('MSE Loss')
ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(f'{OUT_DIR}/01_ae_loss_curve.png', dpi=150); plt.show()

# ─── 2. Reconstruction Error Distribution ───────────────────────────────
re_flat = re_test.flatten()
fig, ax = plt.subplots(figsize=(10, 5))
sns.histplot(re_flat[y_test.values == 0], bins=100, kde=True,
             color='steelblue', label='Normal', stat='density', ax=ax, alpha=0.6)
sns.histplot(re_flat[y_test.values == 1], bins=100, kde=True,
             color='tomato', label='Attack', stat='density', ax=ax, alpha=0.6)
ax.axvline(ae_threshold, ls='--', color='black', lw=2,
           label=f'Threshold P95 = {ae_threshold:.4f}')
ax.set_title(f'AE Reconstruction Error  (MSE ratio = {mse_ratio:.1f}×)',
             fontsize=14, fontweight='bold')
ax.set_xlabel('MSE per Sample'); ax.set_ylabel('Density')
ax.legend(); ax.set_xlim(0, np.percentile(re_flat, 99.5))
plt.tight_layout()
plt.savefig(f'{OUT_DIR}/02_ae_reconstruction_error.png', dpi=150); plt.show()

# ─── 3. Ablation — ROC AUC ──────────────────────────────────────────────
pivot_roc = ablation_df.pivot_table(index='Model', columns='Data', values='ROC AUC')
ax = pivot_roc.plot(kind='bar', figsize=(13, 6),
                    colormap='Set2', edgecolor='black', width=0.7)
ax.set_title('Ablation Study — ROC AUC by Model × Data Variant',
             fontsize=14, fontweight='bold')
ax.set_ylabel('ROC AUC'); ax.set_ylim(0.85, 1.01)
ax.set_xticklabels(ax.get_xticklabels(), rotation=15, ha='right')
ax.legend(title='Data Variant', bbox_to_anchor=(1.01, 1))
for bar in ax.patches:
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.001,
            f'{bar.get_height():.3f}',
            ha='center', va='bottom', fontsize=7)
plt.tight_layout()
plt.savefig(f'{OUT_DIR}/03_ablation_roc.png', dpi=150); plt.show()

# ─── 4. Ablation — F1 Score ─────────────────────────────────────────────
pivot_f1 = ablation_df.pivot_table(index='Model', columns='Data', values='F1')
ax = pivot_f1.plot(kind='bar', figsize=(13, 6),
                   colormap='Set1', edgecolor='black', width=0.7)
ax.set_title('Ablation Study — Weighted F1 by Model × Data Variant',
             fontsize=14, fontweight='bold')
ax.set_ylabel('Weighted F1'); ax.set_ylim(0.85, 1.01)
ax.set_xticklabels(ax.get_xticklabels(), rotation=15, ha='right')
ax.legend(title='Data Variant', bbox_to_anchor=(1.01, 1))
for bar in ax.patches:
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.001,
            f'{bar.get_height():.3f}',
            ha='center', va='bottom', fontsize=7)
plt.tight_layout()
plt.savefig(f'{OUT_DIR}/04_ablation_f1.png', dpi=150); plt.show()

# ─── 5. Ablation Heatmap ────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(16, 5))
for ax, (metric, pivot) in zip(axes, [('ROC AUC', pivot_roc), ('F1', pivot_f1)]):
    sns.heatmap(pivot, annot=True, fmt='.4f', cmap='YlGnBu',
                vmin=0.86, vmax=1.0,
                ax=ax, linewidths=0.5, cbar_kws={'shrink': 0.8})
    ax.set_title(f'Ablation Heatmap — {metric}', fontsize=13, fontweight='bold')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=20, ha='right')
    ax.set_yticklabels(ax.get_yticklabels(), rotation=0)
plt.tight_layout()
plt.savefig(f'{OUT_DIR}/05_ablation_heatmap.png', dpi=150); plt.show()

# ─── 6. All-Model ROC Curves ────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(9, 7))
for i, (mname, (fp, tp, ra)) in enumerate(roc_curves.items()):
    lw = 3 if 'Ensemble' in mname else 1.5
    ls = '-' if 'Ensemble' in mname else '--'
    ax.plot(fp, tp, lw=lw, ls=ls, color=COLORS[i],
            label=f'{mname}  (AUC={ra:.4f})')
ax.plot([0, 1], [0, 1], 'k--', lw=1, alpha=0.5, label='Random')
ax.fill_between(fpr_f, tpr_f, alpha=0.08, color='darkorange')
ax.set_title('ROC Curves — All Models (Full Fusion)',
             fontsize=14, fontweight='bold')
ax.set_xlabel('False Positive Rate'); ax.set_ylabel('True Positive Rate')
ax.legend(loc='lower right', fontsize=10)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(f'{OUT_DIR}/06_roc_all_models.png', dpi=150); plt.show()

# ─── 7. Precision-Recall Curve ──────────────────────────────────────────
fig, ax = plt.subplots(figsize=(9, 6))
ax.plot(rec_f, prec_f, color='darkorange', lw=2.5,
        label=f'Ensemble PR-AUC = {pr_final:.4f}')
ax.fill_between(rec_f, prec_f, alpha=0.15, color='darkorange')
ax.axhline(y_test.mean(), ls='--', color='grey',
           label=f'Baseline (prevalence={y_test.mean():.3f})')
ax.set_title('Precision-Recall Curve — Ensemble (Full Fusion)',
             fontsize=14, fontweight='bold')
ax.set_xlabel('Recall'); ax.set_ylabel('Precision')
ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(f'{OUT_DIR}/07_pr_curve.png', dpi=150); plt.show()

# ─── 8. Confusion Matrix ────────────────────────────────────────────────
tn, fp_cm, fn, tp = cm_final.ravel()
fig, ax = plt.subplots(figsize=(7, 6))
sns.heatmap(cm_final, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Normal', 'Attack'],
            yticklabels=['Normal', 'Attack'],
            annot_kws={'size': 16}, linewidths=1, ax=ax)
ax.set_title(
    f'Confusion Matrix  |  Acc={accuracy_score(y_test, y_pred_f):.4f}'
    f'  |  Threshold={best_thresh:.2f}',
    fontsize=13, fontweight='bold')
ax.set_ylabel('True Label', fontsize=12)
ax.set_xlabel('Predicted Label', fontsize=12)
total = cm_final.sum()
ax.text(0.5, -0.12,
        f'TN={tn} ({tn/total*100:.1f}%)  FP={fp_cm} ({fp_cm/total*100:.1f}%)  '
        f'FN={fn} ({fn/total*100:.1f}%)  TP={tp} ({tp/total*100:.1f}%)',
        ha='center', va='top', transform=ax.transAxes, fontsize=10)
plt.tight_layout()
plt.savefig(f'{OUT_DIR}/08_confusion_matrix.png', dpi=150); plt.show()

# ─── 9. Top-20 Feature Importances ──────────────────────────────────────
# [FIX-7] Use the fitted RF from make_ensemble()
rf_model    = final_model.named_estimators_['rf']
importances = rf_model.feature_importances_
n_feats     = min(len(feat_names), len(importances))
top_idx     = np.argsort(importances[:n_feats])[::-1][:20]
top_names   = [feat_names[i] for i in top_idx]
top_vals    = importances[top_idx]
bar_colors  = ['tomato' if 'ae_' in n or 'latent' in n
                else 'steelblue' for n in top_names]

fig, ax = plt.subplots(figsize=(11, 7))
ax.barh(top_names[::-1], top_vals[::-1],
        color=bar_colors[::-1], edgecolor='white')
ax.set_title('Top-20 Feature Importances (Random Forest)\n'
             'Red = AE/Latent feature   Blue = Original feature',
             fontsize=13, fontweight='bold')
ax.set_xlabel('Importance (Mean Decrease Impurity)')
for i, (name, val) in enumerate(zip(top_names[::-1], top_vals[::-1])):
    ax.text(val + 0.0002, i, f'{val:.4f}', va='center', fontsize=9)
legend_elements = [mpatches.Patch(facecolor='tomato',    label='AE-derived / Latent'),
                   mpatches.Patch(facecolor='steelblue', label='Original')]
ax.legend(handles=legend_elements, loc='lower right', fontsize=11)
plt.tight_layout()
plt.savefig(f'{OUT_DIR}/09_feature_importance.png', dpi=150); plt.show()

# ─── 10. CV Stability ───────────────────────────────────────────────────
# [FIX-7] roc_scores_cv / f1_scores_cv come from the leakage-free loop above
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
for ax, vals, label, color in zip(
        axes,
        [roc_scores_cv, f1_scores_cv],
        ['ROC AUC', 'Weighted F1'],
        ['steelblue', 'seagreen']):
    ax.bar(folds, vals, color=color, edgecolor='black', alpha=0.8)
    ax.axhline(vals.mean(), color='red', ls='--', lw=2,
               label=f'μ={vals.mean():.4f} ± {vals.std():.4f}')
    ax.set_title(f'5-Fold CV (leakage-free) — {label}\nEnsemble / Full Fusion',
                 fontsize=13, fontweight='bold')
    ax.set_xlabel('Fold'); ax.set_ylabel(label)
    ax.set_xticks(folds)
    ax.set_ylim(max(0.90, vals.min() - 0.01), 1.005)
    ax.legend(fontsize=11); ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig(f'{OUT_DIR}/10_cv_stability.png', dpi=150); plt.show()

# ─── 11. PCA Scatter ────────────────────────────────────────────────────
pca   = PCA(n_components=2, random_state=SEED)
idx_s = np.random.choice(len(X_test_fused),
                          min(8000, len(X_test_fused)), replace=False)
X_pca = pca.fit_transform(X_test_fused[idx_s])
y_pca = y_test.values[idx_s]

fig, ax = plt.subplots(figsize=(9, 7))
ax.scatter(X_pca[y_pca == 0, 0], X_pca[y_pca == 0, 1],
           s=8, alpha=0.35, c='steelblue', label='Normal')
ax.scatter(X_pca[y_pca == 1, 0], X_pca[y_pca == 1, 1],
           s=8, alpha=0.35, c='tomato',    label='Attack')
ax.set_title('PCA — Fused Feature Space (test set)',
             fontsize=14, fontweight='bold')
ax.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% var)')
ax.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% var)')
ax.legend(markerscale=4, fontsize=12); ax.grid(True, alpha=0.2)
plt.tight_layout()
plt.savefig(f'{OUT_DIR}/11_pca_scatter.png', dpi=150); plt.show()

# ─── 12. Summary Dashboard (6-panel) ────────────────────────────────────
fig = plt.figure(figsize=(18, 12))
gs  = gridspec.GridSpec(2, 3, figure=fig, hspace=0.42, wspace=0.35)

ax_a = fig.add_subplot(gs[0, 0])
ax_a.plot(history.history['loss'],     lw=2, color=COLORS[0], label='Train')
ax_a.plot(history.history['val_loss'], lw=2, color=COLORS[3], label='Val')
ax_a.set_title('AE Loss Curve', fontweight='bold')
ax_a.set_xlabel('Epoch'); ax_a.set_ylabel('MSE')
ax_a.legend(); ax_a.grid(True, alpha=0.3)

ax_b = fig.add_subplot(gs[0, 1])
re_n = re_test.flatten()[y_test.values == 0]
re_a = re_test.flatten()[y_test.values == 1]
ax_b.hist(re_n, bins=80, density=True, alpha=0.6, color='steelblue', label='Normal')
ax_b.hist(re_a, bins=80, density=True, alpha=0.6, color='tomato',    label='Attack')
ax_b.axvline(ae_threshold, ls='--', color='black', lw=1.5, label='Threshold')
ax_b.set_xlim(0, np.percentile(re_test.flatten(), 99))
ax_b.set_title('Reconstruction Error', fontweight='bold')
ax_b.set_xlabel('MSE'); ax_b.legend()

ax_c = fig.add_subplot(gs[0, 2])
for i, (mname, (fp, tp, ra)) in enumerate(roc_curves.items()):
    lw = 2.5 if 'Ensemble' in mname else 1.2
    ax_c.plot(fp, tp, lw=lw, color=COLORS[i],
              label=f'{mname[:18]}  {ra:.3f}')
ax_c.plot([0,1],[0,1],'k--',lw=1,alpha=0.4)
ax_c.set_title('ROC Curves (Full Fusion)', fontweight='bold')
ax_c.set_xlabel('FPR'); ax_c.set_ylabel('TPR')
ax_c.legend(fontsize=8, loc='lower right')

ax_d = fig.add_subplot(gs[1, 0])
sns.heatmap(cm_final, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Normal','Attack'],
            yticklabels=['Normal','Attack'],
            annot_kws={'size': 14}, ax=ax_d, cbar=False)
ax_d.set_title(f'Confusion Matrix (thr={best_thresh:.2f})', fontweight='bold')
ax_d.set_ylabel('True'); ax_d.set_xlabel('Predicted')

ax_e = fig.add_subplot(gs[1, 1])
ax_e.bar(folds, roc_scores_cv, color='steelblue', edgecolor='black', alpha=0.8)
ax_e.axhline(roc_scores_cv.mean(), color='red', ls='--', lw=2,
             label=f'μ={roc_scores_cv.mean():.4f}')
ax_e.set_ylim(max(0.90, roc_scores_cv.min()-0.01), 1.005)
ax_e.set_title('5-Fold CV ROC AUC (leakage-free)', fontweight='bold')
ax_e.set_xlabel('Fold'); ax_e.legend(fontsize=10)
ax_e.grid(True, alpha=0.3, axis='y')

ax_f = fig.add_subplot(gs[1, 2])
top10_n = [feat_names[i] for i in top_idx[:10]][::-1]
top10_v = importances[top_idx[:10]][::-1]
bc      = ['tomato' if ('ae_' in n or 'latent' in n) else 'steelblue'
           for n in top10_n]
ax_f.barh(top10_n, top10_v, color=bc, edgecolor='white')
ax_f.set_title('Top-10 Feature Importances', fontweight='bold')
ax_f.set_xlabel('Importance')

fig.suptitle(
    f'AE-FIDS — UNSW-NB15 Results Dashboard\n'
    f'ROC AUC={roc_final:.4f}  |  PR AUC={pr_final:.4f}  |  '
    f'CV={roc_scores_cv.mean():.4f}±{roc_scores_cv.std():.4f}  |  '
    f'Threshold={best_thresh:.2f}',
    fontsize=14, fontweight='bold', y=1.01)
plt.savefig(f'{OUT_DIR}/12_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()
print("  Dashboard saved ✓")


# ── CELL 16: Final Summary ───────────────────────────────────────────────
summary = f"""
╔══════════════════════════════════════════════════════════════════╗
║             AE-FIDS  —  FINAL RESULTS SUMMARY (FIXED)            ║
╠══════════════════════════════════════════════════════════════════╣
║ Dataset        : UNSW-NB15                                       ║
║ Train samples  : {len(y_train):>10,}  (50/50 balanced)               ║
║ Test  samples  : {len(y_test):>10,}  (original distribution)         ║
╠══════════════════════════════════════════════════════════════════╣
║ BUGS FIXED                                                       ║
║  [FIX-1] Train/test file names were swapped → corrected          ║
║  [FIX-2] CV leakage: AE + scaler + selector now re-fitted        ║
║          inside each fold on fold-train data only                ║
║  [FIX-3] ens_cv variable was undefined → replaced with           ║
║          roc_scores_cv / f1_scores_cv from the leakage-free loop ║
║  [FIX-4] Normal recall improved: decision threshold tuned        ║
║          (sweep 0.30–0.70, constrain normal recall ≥ 0.70)       ║
║  [FIX-5] Latent vector (8-dim) added to fused features →         ║
║          Full Fusion now strictly dominates No AE baseline        ║
║  [FIX-6] XGBoost + LightGBM added as stronger baselines          ║
║  [FIX-7] roc_folds / f1_folds now correctly reference the        ║
║          leakage-free CV arrays                                  ║
║  [FIX-8] matplotlib.patches imported for feature importance plot ║
╠══════════════════════════════════════════════════════════════════╣
║ AUTOENCODER                                                      ║
║   Architecture : {input_dim}→128→64→32→{encoding_dim}→32→64→128→{input_dim}                      ║
║   Epochs run   : {epochs_run:<6}                                        ║
║   Val MSE loss : {val_loss:.6f}                                   ║
║   MSE ratio    : {mse_ratio:.2f}×  (attack / normal)                  ║
║   Threshold    : {ae_threshold:.6f}  (P95 of normal train MSE)     ║
╠══════════════════════════════════════════════════════════════════╣
║ BEST MODEL: Ensemble (RF + HGB [+ XGB + LGB]) — Full Fusion      ║
║   Decision threshold : {best_thresh:.2f}                               ║
║   ROC AUC      : {roc_final:.4f}                                      ║
║   PR  AUC      : {pr_final:.4f}                                      ║
║   CV ROC AUC   : {roc_scores_cv.mean():.4f} ± {roc_scores_cv.std():.4f}  (leakage-free)     ║
║   CV F1        : {f1_scores_cv.mean():.4f}  ± {f1_scores_cv.std():.4f}  (leakage-free)     ║
╠══════════════════════════════════════════════════════════════════╣
║ SAVED FILES                                                      ║
║   autoencoder.keras   ablation_results.csv  final_summary.txt   ║
║   X_train/test_fused.npy   01_ae_loss.png … 12_dashboard.png    ║
╚══════════════════════════════════════════════════════════════════╝
"""
print(summary)
with open(f'{OUT_DIR}/final_summary.txt', 'w') as f:
    f.write(summary)

print("ALL DONE ✓  — check /kaggle/working/ for outputs")
```
